#### **Actualizando un banco de datos**

```python
query = 'SELECT * FROM clientes'
```

```python
pd.read_sql(sql=text(query), con=engine.connect())
```

```python
#Borrando registros.
from sqlalchemy.exc import SQLAlchemyError
querry = "DELETE FROM clientes WHERE ID_Cliente=5008804"
try:
  r_set = engine.connect().execute(text(querry))
except SQLAlchemyError as e:
  print(e)
else:
  print("#Registros borrados", r_set.rowcount)
```

```python
pd.read_sql_table('clientes', con=engine.connect())
```

```python
#Actualizando registros.
querry = "UPDATE clientes SET Grado_estudio='Nivel superior' WHERE ID_Cliente=5008808"
try:
  r_set = engine.connect().execute(text(querry))
except SQLAlchemyError as e:
  print(e)
else:
  print("#Registros actualizados", r_set.rowcount)
```

```python
pd.read_sql_table('clientes', con=engine.connect())
```
